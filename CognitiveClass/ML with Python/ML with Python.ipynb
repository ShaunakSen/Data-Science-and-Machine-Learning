This function produces one output and requires two inputs.
<ul>
    <li> 1st Input: A pandas array. The pandas array we have been using is my_data </li>
    <li> 2nd Input: Any number of integer values (order doesn't matter) that represent the columns that we want to remove. We also want to remove the first column because that only contains the row number, which is irrelevant to our analysis.</li>

    </ul>
</ul>


```python
my_data.columns
```

```python
# Remove the column containing the target name since it doesn't contain numeric values.
# Also remove the column that contains the row number
# axis=1 means we are removing columns instead of rows.
# Function takes in a pandas array and column numbers and returns a numpy array without
# the stated columns


def removeColumns(pandasArray, *column):
    
    # print ([column])
    # print (pandasArray.values)
    
    return pandasArray.drop(pandasArray.columns[[column]], axis=1).values
    
new_data = removeColumns(my_data, 0,1)

print(new_data)
```

Now, we have one half of the required data to fit a model, which is X or new_data

Next, we need to get the response vector y. Since we cannot use .target and .target_names, I have created a function that will do this for us.


<b> targetAndtargetNames(numpyArray, targetColumnIndex) </b>

This function produces two outputs, and requires two inputs.
<ul>
    <li> <font size = 3.5><b><i>1st Input</i></b></font>: A numpy array. The numpy array you will use is my_data.values (or my_data.as_matrix())</li>
    <ul>
        <li> Note: DO NOT USE <b> new_data </b> here. We need the original .csv data file without the headers </li>
    </ul>
</ul>
<ul>
    <li> <font size = 3.5><b><i>2nd Input</i></b></font>: An integer value that represents the target column . (Look at the data again and find which column contains the non-numeric values. This is the target column)</li>
    <ul>
        <li> Note: Remember that Python is zero-indexed, therefore the first column would be 0. </li>
   </ul>
</ul>

<ul>
    <li> <font size = 3.5><b><i>1st Output</i></b></font>: The response vector (target) </li>
    <li> <font size = 3.5><b><i>2nd Output</i></b></font>: The target names (target_names) </li>
</ul>



```python
def targetAndtargetNames(numpyArray, targetColumnIndex):
    target_dict = dict()
    target = list()
    target_names = list()
    count = -1
    for i in range(len(my_data.values)):
        if my_data.values[i][targetColumnIndex] not in target_dict:
            count += 1
            target_dict[my_data.values[i][targetColumnIndex]] = count
        target.append(target_dict[my_data.values[i][targetColumnIndex]])
    # Since a dictionary is not ordered, we need to order it and output it to a list so the
    # target names will match the target.
    for targetName in sorted(target_dict, key=target_dict.get):
        target_names.append(targetName)
    return np.asarray(target), target_names

target, target_names = targetAndtargetNames(my_data, 1)

print(target)

print(len(target))

print (my_data.values)

print(target_names)
```

#### Random Forests

RF: many DTs

Each DT is created from random splits in the dataset


**RF Algo**

1. For b=1 to B:

    - Draw a bootstrap sample Z* of size N from the training data
    
    - Grow a RF tree Tb to the bootstrapped data, by recursively repeating the following steps for each terminal
    node of the tree, until the minm node size n<sub>min</sub> is reached
    
        - Select m vars randomly from p vars
        
        - Pick best var/split pt among the m
        
        - Split the node into 2 daughter nodes
    
2. Output the ensemble of trees {Tb}<sub>1</sub><sup>B</sup>

B: no of trees we wanna create

Bootstrapping : ** u randomly select N data pts (with replacement) from a set of data**

Some data pts may be chosen more than once and some may not be chosen at all

2 sources of randomness:

1. Randomness in the data

2. Randomness in the Feature splits



#### Why RF when we have DTs already

DTs are easy to build, use and interpret

new samples**

RF combines simplicity of DTs with flexibility resulting in a vast improvement in accuracy

Also another main reason is **Size of Dataset**:

This may not be true for other classifiers like KNN

** Bagging : Bootstrapping the data plus using the aggregrate to make a decision is called Bagging **

This allows RF to be a v good estimator

**K-Nearest Neighbors** is an algorithm for supervised learning. Where the data is 'trained' with data points corresponding to their classification. Once a point is to be predicted, it takes into account the 'K' nearest points to it to determine it's classification.

In this sense, it is important to consider the value of k. But hopefully from this diagram, you should get a sense of what the K-Nearest Neighbors algorithm is. It considers the 'K' Nearest Neighbors (points) when it predicts the classification of the test point.

#### Train and test on same dataset:

we train model on dataset and test it on same dataset

This evaluation model will likely have a high training accuracy and a low out-of-sample accuracy, since the model **knows all the testing datapts from the training**

What exactly is **training accuracy**

- training accuracy is the % of correct predns that the model makes when using the training dataset

However, a high training accuracy isnt necessarily a good thing

High training accuracy may result in **Overfitting**

This means that the model is overly trained to the dataset which may capture noise and produce a non generalized model

Out of Sample accuracy: % of correct predns that model makes on data that model has not been trained on

Doing train and test on same dataset will likely have a low OoS accuracy because of being overfit

High Out of Sample accuracy is imp as purpose of model is to have it perform well on unknown data

#### Train test split

Improves OoS accuracy

Data is split into train and test sets

Train with training set, test with testing set

This provides a more accurate eval on OoS accuracy as test data is part of daatset that could have been used to train the model

This means that we know outcome of each data pt in the dataset making it great to test with.And since this data
has not been used to train the model,it has no knowledge of the outcome of these data pts

So,in essence its truly an OoS testing

Ensure u train ur model with test data afterwards so as not to lose potentially valuable data

Prob : which dataset is used for train and which for test
If test set has too many outliers we will get a bad estimate for OoS accuracy

Soln: k fold cross validation

Then result is avged to produce a more consistent OoS accuracy

### Overfitting and Underfitting

**Bias**

- This is the error that results from incorrect assumptions and relations that the model makes about the data. In stats and prob theory bias occurs when **the expected value of an estimator differs from the true underling parameter which is being estimated**

- High bias tend to create **Overly generalized** models which can cause a loss of relevant relations bw the features and target op. 

- High bias == Underfitting

**Variance**

- This is the inconsistency of the model due to small changes in the dataset

- In prob and stats theory, variace is the **expected value of the squared deviation of a random variable from its mean**. It can also be defined as the **measure of the spread bw the numbers in a dataset and its mean**

- High var == Overfitting

#### Bias - Variance Trade off

Errors in bias and variance are defined as the **diff in expected predn** and **variability of model predn** for a given data pt respectively

Since bias and variance are 2 ends of the spectrum,one causing overfittinga nd the otherunderfitting we have to keep them at a balance

Overly complex model -> Overfit -> Not generalied to OoS data

A good balance is keeping the model general enough for OoS data but specific enough to **fit the pattern of th data**
