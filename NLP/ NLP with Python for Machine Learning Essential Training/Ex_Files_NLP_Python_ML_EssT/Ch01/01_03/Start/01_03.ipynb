# NLP Basics: Reading in text data & why do we need to clean the text?

### Read in semi-structured text data

```python
# Read in the raw text
rawData = open("SMSSpamCollection.tsv").read()

# Print the raw data
rawData[0:500]
```

```python
parsedData = rawData.replace('\t', '\n') # replace tabs with new lines

parsedData[:500]
```

```python
parsedData = parsedData.split('\n')

parsedData[:10]
```

```python
labelList = parsedData[0::2] # start in 0th pos go till very end and take every 2nd item

labelList
```
