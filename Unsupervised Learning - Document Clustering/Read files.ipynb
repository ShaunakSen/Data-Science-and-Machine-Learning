### Create a text processing pipeline

[link](https://machinelearningmastery.com/clean-text-machine-learning-python/)

```python
import string
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
```

```python
from nltk.tokenize import word_tokenize
def process_text(text):
    # split into words
    tokens = word_tokenize(text)
    # convert to lower case
    tokens = [w.lower() for w in tokens]
    # remove punctuation from each word
    table = str.maketrans('', '', string.punctuation)
    stripped = [w.translate(table) for w in tokens]
    # remove remaining tokens that are not alphabetic
    words = [word for word in stripped if word.isalpha()]
    # filter out stop words
    stop_words = set(stopwords.words('english'))
    words = [w for w in words if not w in stop_words]
    return " ".join(words)
```

```python
main_text = []

# should be of len 24
```

```python
def get_text_from_file(filepath):
    html_doc = open(filepath).read()
    soup = BeautifulSoup(html_doc, 'html.parser')
    all_text_tags = soup.find_all('span', attrs={'class': 'ocr_cinfo'})
    all_text_words = []

    for text_tag in all_text_tags:
        text_in_tag = text_tag.text.strip()
        if (len(text_in_tag) > 1):
            processed_text = process_text(text_in_tag)
            all_text_words.append(processed_text)
        
    return " ".join(all_text_words)
```

```python
def get_text_from_folder(list_of_files, index):
    folder_text = []
    folder_name = all_folders[index]
    for file_name in list_of_files:
        file_path = folder_name + '/' + file_name
        folder_text.append(get_text_from_file(file_path))
        
    main_text.append(folder_text)
        
```

```python
for x in range(len(all_folders)):
    print("Folder name", all_folders[x])
    all_files_in_folder = files_in_each_folder[x]
    get_text_from_folder(all_files_in_folder, x)
```

```python
for index in range(len(main_text)):
    print (index)
    text = ' '.join(main_text[index])
    file_name = str(index) + '.txt'
    file = open(file_name, "w")
    file.write(text)
    file.close()
    
```
