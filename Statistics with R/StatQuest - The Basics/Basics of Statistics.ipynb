### What is a statistical distribution?

Imagine we measure heights of a bunch of people

We get a histogram :

Something like this

![](./data/img/diag1.png)

1. Most people are bw 5 and 6 feet tall

2. People smaller than 5ft or >6ft are rare

There is high prob that a random guy will be bw 5 to 6 ft

![](./data/img/diag2.png)


By measuring more people and using smaller bins we get more accurate and more precise estimate as to how hts are distributed

But if we make bin sizes too small, each measurement almost will get its own bin, so it wont be useful anymore

We can use a curve to approx the hist

![](./data/img/diag3.png)


But this curve has a few adv over histogram:

Even though for a particular bin we dont have a measurement, we can use the curve to calculate prob 

![](./data/img/diag4.png)

Another adv is that the curve is not limited by the width of the bins

![](./data/img/diag5.png)

![](./data/img/diag6.png)


Both hist and the curve are distributions as they show us how the probs of measurements are distributed

### Sampling A Distribution

We already know what a ND is

Now we want to know what it means to take a sample from a dist

![](./data/img/diag10.png)

![](./data/img/diag11.png)

However, every now and then, computer will return a value from the edges where the histogram and curves are shortest

![](./data/img/diag12.png)


For eg,

We can take 2 samples (N=3) from a single dist and do T tests on the samples
N=3: we take 3 measurements from each sample

Since the dist is same, the T test should give a large p value

Doing lots of tests will give me a sense of how frequently the T test successully gives me a large p value

If i had 2 separate dists a T test is supposed to give me a small p value

![](./data/img/diag13.png)

Taking sample from a dist or multiple dists, ie getting a comp to generate random numbers that reflect the prob of the dist, let us determine what a stat test is capable of doing without much effort




### The Central Limit Theorem

Lets start with a **Uniform Distribution**

There is an equal prob of selecting values bw 0 and 1

![](./data/img/diag14.png)


We collect 20 random samples from this dist

![](./data/img/diag15.png)

Then we calculate the mean of these 20 samples (0.55) and on the right we plot this mean on a histogram

We do this process again and again...

![](./data/img/diag16.png)

![](./data/img/diag17.png)

![](./data/img/diag18.png)

The means are calculated using data from a **Uniform Dist**

But these means are **not uniformly distributed. The means are Normally distributed**

---

Another eg

We start with an **exponential dist**

We collect 20 random samples, calculate mean and plot it on a hist

![](./data/img/diag19.png)

![](./data/img/diag20.png)


The means are calculated using data from an **Exp Dist**

But these means are **not exp distributed. The means are Normally distributed**

It does not matter what dist we start with, if we collect samples from those dist, the means will be Normally dist (some conditions apply but we wont worry about that too much)
