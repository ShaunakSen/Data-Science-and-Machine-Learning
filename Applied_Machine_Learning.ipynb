### Figure and axes

figure = one window or one image file

axes = drawing areas with coordinate system

by default: each figure has one axis

But very often we might want more than 1 axes for a figure

![](https://amueller.github.io/COMS4995-s20/slides/aml-02-matplotlib/images/figure_axis.png)


#### Creating figures and axes

- `fig = plt.figure()`
    Creates a figure with axes, sets current figure. Can add more / different axes later.

- `fig, ax = plt.subplots(n, m)`

Creates a figure with a regular grid of n x m axes.

![](https://amueller.github.io/COMS4995-s20/slides/aml-02-matplotlib/images/subplots.png)

See more: https://matplotlib.org/stable/gallery/subplots_axes_and_figures/subplots_demo.html



```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# Some example data to display
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('A single plot')
```

```python
fig, axs = plt.subplots(2)
fig.suptitle('Vertically stacked subplots')
axs[0].plot(x, y)
axs[1].plot(x, -y)
```

```python
fig, ax = plt.subplots(1, 2)
fig.suptitle('Horizontally stacked subplots')
ax[0].plot(x, y)
ax[1].plot(x, -y)
```

When stacking in two directions, the returned axs is a 2D NumPy array.

If you have to set parameters for each subplot it's handy to iterate over all subplots in a 2D grid using for ax in axs.flat:.

```python
fig, axs = plt.subplots(2, 2)
axs[0, 0].plot(x, y)
axs[0, 0].set_title('Axis [0, 0]')
axs[0, 1].plot(x, y, 'tab:orange')
axs[0, 1].set_title('Axis [0, 1]')
axs[1, 0].plot(x, -y, 'tab:green')
axs[1, 0].set_title('Axis [1, 0]')
axs[1, 1].plot(x, -y, 'tab:red')
axs[1, 1].set_title('Axis [1, 1]')

for ax in axs.flat:
    ax.set(xlabel='x-label', ylabel='y-label')

# Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs.flat:
    ax.label_outer()
```
