### Import the libraries



```
import numpy as np
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
```

### Get the data

The data we are going to use is from John Hopkin Uni - [link](https://github.com/CSSEGISandData/COVID-19/tree/master/csse_covid_19_data/csse_covid_19_time_series)

These are live links and the data will be updated regularly



```
confirmed_cases_file_link = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
death_cases_file_link = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv'
recovered_cases_file_link = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv'
country_cases_file_link = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/web-data/data/cases_country.csv'
```

```
confirmed_df = pd.read_csv(confirmed_cases_file_link)
print (confirmed_df.shape)
confirmed_df.head()
```

```
deaths_df = pd.read_csv(death_cases_file_link)
print (deaths_df.shape)
deaths_df.head()
```

```
recovered_df = pd.read_csv(recovered_cases_file_link)
print (recovered_df.shape)
recovered_df.head()
```

```
cases_country_df = pd.read_csv(country_cases_file_link)
print (cases_country_df.shape)
cases_country_df.head()
```

The level of data is `Province/State-Country/Region`	

```
confirmed_df.loc[confirmed_df['Country/Region']=='Australia']
```

Are all the countries in the datasets same?

```
countries_confirmed, countries_deaths, countries_recovered, countries_cases = set(confirmed_df['Country/Region'].unique()), set(deaths_df['Country/Region'].unique()), set(recovered_df['Country/Region'].unique()), set(cases_country_df['Country_Region'].unique())
```

```
print (len(countries_confirmed), len(countries_deaths), len(countries_recovered), len(countries_cases))
```

```
print (countries_confirmed - countries_deaths)
print (countries_confirmed - countries_recovered)
print (countries_confirmed - countries_cases)
```

> So, all the countries are present in all the dfs

### Imputing the data

```
confirmed_df = confirmed_df.replace(to_replace=np.nan, value='', regex=True)
deaths_df = deaths_df.replace(to_replace=np.nan, value='', regex=True)
recovered_df = recovered_df.replace(to_replace=np.nan, value='', regex=True)
cases_country_df = cases_country_df.replace(to_replace=np.nan, value='', regex=True)
```
